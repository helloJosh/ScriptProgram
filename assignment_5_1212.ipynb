### [2번 문제]
#### 다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.

다형성이란 동일한 객체의 동일한 메소드가 여러 종류의 파라미터를 받고 그에 대응하는 여러 종류의 결과값을 낼 수 있는 기능을 구현하는 것이다. 따라서 파이썬에서는 변수의 타입이 동적으로 결정되므로 다형성을 구현하기가 힘들다. 같은 이름의 메소드여도 다른 파라미터를 받으면 각각 원하는 결과를 이끌어 낼 수 있어야한다.

```python
class wrapper:
    def __init__(self, data):
        self.data = data
        
class Float_toString(wrapper):
    def toString(self):
        if type(self.data) is float:
            return str(self.data)
        
class Int_toString(wrapper):
    def toString(self):
        if type(self.data) is int:
            return str(self.data)

for i in (Float_toString(3.1), Int_toString(3)):
    if isinstance(i, wrapper):
        print(i.toString())
```

파이썬 특유의 유연한 구조는 엄격한 타입 검사를 수행하는 코드를 구현하려면 되려 기존보다 코드가 늘어나는 단점있다.
다형성을 표현하기 위해서 3개의 클래스를 정의했다. wrapper는 데이터를 포장하는 클래스이고, Float_toString과 Int_toString는 wrapper를 상속하는 클래스이다. Float_toString과 Int_toString 메소드를 구현한다. 같은 이름이지만 동작은 다르다.

출력 부분의 코드를 보면 for 반복문에 넘어간 튜플 내의 모든 원소가 wrapper 클래스의 인스턴스라는 것을 확인할 수 있다. 이 for 반복문은 튜플의 각 원소에 대해 toString을 수행한다. 단형성 언어라면 toString() 메서드의 결과는 모두 똑같아야 한다. 
하지만 결과는 float일때와 int일 때와 달랐다. 이러한 것이 다형성이다.

### [3번 문제]
#### 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

    - step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)
    
>&gt;&gt;&gt; c = Counter(10)  
>&gt;&gt;&gt; d = Counter(10, 2) 

- 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.

- 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

>&gt;&gt;&gt; c.incr()

>[Count (step: 1)] 11

- 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.

>&gt;&gt;&gt; c()   
>&gt;&gt;&gt; d()   
>&gt;&gt;&gt; print(c)   
>[Count (step: 1)] 12   
>&gt;&gt;&gt; print(d)   
>[Count (step: 2)] 14   

- 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

>&gt;&gt;&gt; c = c + 5   
>&gt;&gt;&gt; d = d - 5   
>&gt;&gt;&gt; print(c)   
>[Count (step: 1)] 17   
>&gt;&gt;&gt; print(d)   
>[Count (step: 2)] 9   

- 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

>&gt;&gt;&gt; print(c) > 10   
True   
>&gt;&gt;&gt; print(d) > 10   
False   
>&gt;&gt;&gt; print(c) < 10   
False   
>&gt;&gt;&gt; print(d) < 10   
True   
>&gt;&gt;&gt; print(c) == 17   
True   
>&gt;&gt;&gt; print(d) != 9   
False   

```python
class Counter():
    # 요구사항 1 구현
    def __init__(self, num, step = 1):
        self.num = num
        self.step = step
        
    # 요구사항 2 구현
    def __str__(self):
        return "[Count (step: " + str(self.step) + ")] " + str(self.num)
    
    # 요구사항 3 구현
    def incr(self):
        self.num += self.step
    
    # 요구사항 4 구현
    def __call__(self):
        self.incr()
    
    # 요구사항 5 구현
    def __add__(self, other):
        return self.num + other
    
    def __sub__(self, other):
        return self.num - other
    
    # 요구사항 6 구현
    def __cmp__(self, other):
        if self.num < other.num :
            return -1
        elif self.num == other.num :
            return 0
        else :
            return 1
```

 __ 으로 시작하고 끝나는 메소드명은 예약어로, 해당 클래스 타입에 연산자 오버로딩을 추가하거나, 특정 처리에 대응하여 처리할 코드를 추가하기 위해 사용한다.

```python
c = Counter(10)
d = Counter(10, 2)

#요구사항 2
print("요구사항 2")
print(c)
print(d)

#요구사항 3
print("\n요구사항 3")
c.incr()
d.incr()
print(c)
print(d)

#요구사항 4
print("\n요구사항 4")
c()
d()
print(c)
print(d)

#요구사항 5
print("\n요구사항 5")
c = c + 5
d = d - 5
print(c)
print(d)

#요구사항 6
print("\n요구사항 6")

print(c > 10)
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 9)
```

### [4번 문제]
#### 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + ", "
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
```

```python
s = MySet([1, 2, 2, 3])
print(s)
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
print(t)
```

__init__ 는 생성자 코드로, 임의의 리스트 l을 파라미터로 받는다. MySet의 인스턴스가 생성될 때, l의 모든 원소를 자신에게 추가한 뒤 eliminate_duplicate를 호출한다. 

__str__ 는 print, 혹은 str() 등 객체를 문자열화 할 필요가 있을 때 수행되는 메서드이다.

eliminate_duplicate는 리스트 내의 각 요소에 대해 중복되는 요소를 제거하는 역할을 한다. 새로운 리스트를 만들고, 이 리스트에는 MySet의 인스턴스가 갖고 있는 리스트 원소를 추가한다. 그러나 만약 복사하려는 원소를 이미 포함하고 있다면 추가하지 않는다. 이후에 MySet의 인스턴스가 갖고 있는 리스트를 초기화하고, 중복 제거된 리스트의 원소들을 다시 추가한다.

### [5번 문제]
#### 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

- 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

>&gt;&gt;&gt; u = s | t   
>&gt;&gt;&gt; print(u)   
>MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}   

- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

>&gt;&gt;&gt; u = s & t   
>&gt;&gt;&gt; print(u)   
>MySet: {2, 3}   

- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

>&gt;&gt;&gt; s = MySet([1, 2, 3])   
>&gt;&gt;&gt; t = MySet([3, 4, 5])   
>&gt;&gt;&gt; u = s - t   
>&gt;&gt;&gt; print(u)   
>MySet: {1, 2}   

```python
class MySet(list):
    # 원본 코드
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + ", "
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    # 요구사항 1 구현
    def __or__(self, l):
        s = []
        for e in l + self:
            s.append(e)
        MySet.eliminate_duplicate(s)
        return MySet(sorted(s))
    # 요구사항 2 구현
    def __and__(self, l):
        s = []
        for e in l:
            if e in self:
                s.append(e)
        return MySet(s)
    
    # 요구사항 3 구현
    def __sub__(self, l):
        s = []
        for e in self:
            if not e in l:
                s.append(e)         
        return MySet(s)
```

```python
u = s or t
print(u)

#요구사항 2
u = s and t
print(u)

#요구사항 3
s = MySet([1, 2, 3])
t = MySet([3, 4, 5])
u = s - t
print(u)
```

 __or__ 메서드는 self, l 파라미터를 받아 이 두 리스트의 모든 요소를 s에 넣고, s에 대해 중복제거를 수행한 뒤 정렬한다. 이 리스트를 이용해 MySet 인스턴스를 만들어 리턴한다. 즉, 합집합을 구현한다.

__and__ 메서드는 self, l 파라미터를 받아 l 리스트를 돌면서 self에도 해당 요소가 있으면 s에 추가한다. 즉, 교집합을 구현한다.

__sub__ 메서드는 self 의 요소를 순회하면서 l에 해당 요소가 없는 경우에만 s에 추가한다. 즉, 차집합을 구현한다.

```python
import pickle
import urllib.request
import string
from operator import itemgetter

class SearchEngine:
    #전 assignment 4 코드 delete_code, project
    def delete_Code(self , source, start, end):
        deleted_Code = ""
        s_l = source.split(start)

        for i in range(0, len(s_l)):
            sub = s_l[i]
            index = sub.rfind(end)
            if(index == -1 or index == len(sub)-1):
                if( i != 0 and i != len(s_l)-1):
                    continue
            deleted_Code += sub[index+1:] + " "
        deleted_Code = " ".join(deleted_Code.split())
        return deleted_Code
    
    def assignment4(self, url):
        #전에 했던 urllib가 갑자기 작동하지 않아 수정하였습니다.
        response = urllib.request.urlopen('http://python.org/')
        source = response.read().decode("utf8")
        
        source = self.delete_Code(source, '<script', '/script>')
        source = self.delete_Code(source, '<style', '/style>')
        source = self.delete_Code(source, '<', '>')
        for i in range(0, len(string.punctuation)):
            source = source.replace(string.punctuation[i], "")

        words = {}
        for i in source.split():
            if(i in stopWords):
                continue
            if(not i in words.keys()):
                words[i]=1
            else:
                words[i]=words[i]+1
        return words

    # url을 임의의 개수 넣을 수 있는 생성자 정의
    def __init__(self, *arg):
        self.urlList = []
        for i in arg:
            if not i in self.urlList:
                self.urlList.append(i)
    
    # addUrl을 이용해 url을 단 하나 추가할 수 있음
    def addUrl(self, url):
        if not url in self.urlList:
            self.urlList.append(url)
    
    # removeUrl을 이용해 url을 단 하나 삭제할 수 있음
    def removeUrl(self, url):
        if url in self.urlList:
            self.urlList.remove(url)
    
    # listUrls()
    def listUrls(self):
        for url in self.urlList:
            print(url)
    
    # getWordsFrequency 
    def getWordsFrequency(self):
        words = {}
        for url in self.urlList:
            wordsInUrl = self.assignment4(url)
            for i in wordsInUrl.keys():
                if not i in words.keys():
                    words[i] = wordsInUrl[i]
                else:
                    words[i] += wordsInUrl[i]
        return words

    # getMaxFreqencyWords 
    def getMaxFreqencyWords(self):
        words = self.getWordsFrequency()
        
        if not len(words.items()) == 0:
            wordList = []
            sortedTupleList = sorted(words.items(), key=itemgetter(1), reverse=True)
            freq = sortedTupleList[0][1] 
            i = 0
            while sortedTupleList[i][1] == freq:
                wordList.append(sortedTupleList[i][0])
                i+=1
                if i == len(sortedTupleList):
                    break
            return wordList
        else:
            return None
        
    # searchUrlByWord
    def searchUrlByWord(self, word):
        wordsList = []
        max = -1
        for url in self.urlList:
            wordsInUrl = self.assignment4(url)
            if word in wordsInUrl.keys():
                wordsList.append((url, wordsInUrl[word]))
                if wordsInUrl[word] > max:
                    max = wordsInUrl[word]
        if not max == -1:
            for s in wordsList:
                if max == s[1]:
                    print(s[0])
        else:
            print("there is no " + word)
```

### [8번 문제]
#### 위 7번 문제에서 정의한 SearchEngine 클래스를 상속하여 SearchEngineWithOrderedWebWords 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

- 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
- getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.

    - reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.
    - reverse 인자에 true를 넣으면 오름차순으로 정렬한다.
    
    >&gt;&gt;&gt; w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
    >  
    >&gt;&gt;&gt; w4.getWordsFrequency()  
    >[('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]  
    >  
    >&gt;&gt;&gt; w4.getWordsFrequency(reverse=true)  
    >[('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)  

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    # 8번 문제
    def getWordsFrequency(self, reverse=False):
        w1 = SearchEngine()
        for url in self.urlList:
            w1.addUrl(url)
        words = w1.getWordsFrequency()
        sortedwords = sorted(words.items(), key=itemgetter(1), reverse=not reverse)
        return sortedwords
    
    # 9번 문제
    def __iter__(self):
        return iter(self.getWordsFrequency())
```
